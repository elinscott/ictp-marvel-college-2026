# Extracting the IP and EA from the koopmans output

This notebook loads the `ozone.pkl` file generated by `koopmans` and prints the ionisation potential (IP) and electron affinity (EA). Run all the cells and check that you get the same numbers as in Problem 5 Part A.

```python
from koopmans import io

# Load the workflow object
wf = io.read('ozone.pkl')
```

```python
# Access the results from the very last calculation (the final KI calculation)
results = wf.calculations[-1].results

# The IP and EA are minus the HOMO and LUMO eigenvalues
ip = -results['homo_energy']
ea = -results['lumo_energy']

print(f' IP = {ip:.2f} eV')
print(f' EA = {ea:.2f} eV')
```
